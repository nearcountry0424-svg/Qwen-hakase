```python
%cd /notebooks/ComfyUI    

import os
import time
from threading import Thread
import importlib
import subprocess

def run_comfyui():
    try:        
        os.system("/notebooks/ComfyUI/venv/bin/python main.py --listen 0.0.0.0 --port 6006 --preview-method auto")
    except Exception as e:
        if "torch.OutOfMemoryError: Allocation on device" in str(e):
            print(f"警告: GPUメモリ不足エラー: {str(e)}")
        elif "RuntimeError: CUDA out of memory" in str(e) or "RuntimeError: DefaultCPUAllocator: not enough memory" in str(e) or "MemoryError" in str(e):
            print(f"警告: メモリ不足エラー: {str(e)}")
        else:
            print(f"ComfyUI実行エラー: {str(e)}")
            
def restart_comfyui():
    os.system("pkill -f main.py")
    time.sleep(3)
    
    fqdn = os.environ.get('PAPERSPACE_FQDN')
    if fqdn:
        localurl = f"https://tensorboard-{fqdn}"
        print(f"=====================================================================")
        print(f"公開URL: {localurl}")
        print(f"※ 後ほど、http:// 0.0.0.0:6006 のURLが表示されますので、その後「公開URL:（ {localurl} ）」の画面を更新してください。")
        print("※ なお、http:// 0.0.0.0:6006 のURLの方はクリックしないようご注意ください。")
        print(f"=====================================================================")
    else:
        print("FQDN未設定")

    try:
        python_version_result = subprocess.run(["/notebooks/ComfyUI/venv/bin/python", "--version"], capture_output=True, text=True, check=True)
        python_version_result.stdout.strip()
        # print(f"Python : {python_version_result.stdout.strip()}")
    except (subprocess.CalledProcessError, FileNotFoundError) as e:
        # print(f"Python バージョン確認でエラーが発生しました: {e}")
        %cd /notebooks
        !apt-get update -y
        !apt-get install python3.12 python3.12-venv python3.12-dev build-essential -y
        %cd /notebooks/ComfyUI    

    pip_executable = "/notebooks/ComfyUI/venv/bin/pip"
    pip_options = [
        "--extra-index-url", "https://download.pytorch.org/whl/cu121",
    ]

    def install_package(package, pip_executable, pip_options):
        try:
            subprocess.check_call([pip_executable, "install", package] + pip_options)
        except subprocess.CalledProcessError as e:
            print(f"エラー: {package} インストール失敗 {e}")

    requirements_path = "/notebooks/ComfyUI/requirements.txt"
    
    if os.path.exists(requirements_path):
        print(f"基本requirements.txtからインストール中...")
        try:
            !{pip_executable} install -r {requirements_path}
            print("基本requirements.txtインストール完了")
        except Exception as e:
            print(f"警告: 基本requirements.txtインストール失敗 {e}")
    else:
        print(f"警告: 基本requirements.txt ({requirements_path}) が見つかりません")

    specific_packages = [
        "sageattention==1.0.6", "google-generativeai", "sounddevice", "gguf",
        "fal_client", "pytube", "GPUtil", "openai",  "anthropic==0.54.0", 
        "mediapy", "rotary_embedding_torch", "aiofiles", "ultralytics",  "wget", 
        "gdown", "surrealist", "boto3",  "albumentations", "numba", "langdetect",  "translators", 
        "wget",  "ultralytics", "pytube",  "hoge", "hoge",  "hoge", "hoge",  "hoge", "hoge",  "hoge", 
        "https://github.com/nunchaku-tech/nunchaku/releases/download/v0.3.2dev20250715/nunchaku-0.3.2.dev20250715+torch2.6-cp312-cp312-linux_x86_64.whl",
    ]
    try:
        !{pip_executable} install {' '.join(specific_packages)}
    except Exception as e:
        print(f"警告: 特定パッケージインストール失敗 {e}")

    thread = Thread(target=run_comfyui)
    thread.daemon = True
    thread.start()
    time.sleep(5)
    print("ComfyUI再起動完了")

import os
import matplotlib
os.environ['MPLBACKEND'] = 'Agg'
matplotlib.use('agg') 
import matplotlib.pyplot as plt
print(f"Matplotlib backend set to: {matplotlib.get_backend()}")

try:
    restart_comfyui()
except Exception as e:
    if "torch.OutOfMemoryError: Allocation on device" in str(e):
        print(f"警告: GPUメモリ不足エラー: {str(e)}")
    elif "RuntimeError: CUDA out of memory" in str(e) or "RuntimeError: DefaultCPUAllocator: not enough memory" in str(e) or "MemoryError" in str(e):
        print(f"警告: メモリ不足エラー: {str(e)}")
    else:
        print(f"エラー発生: {str(e)}")
```
